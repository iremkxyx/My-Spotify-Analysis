#### ts: Bu alan, UTC (Koordinatlı Evrensel Saat) zamanıyla  parçanın ne zaman çalmayı bitirdiğini gösteren zaman damgasıdır.  Sırasıyla yıl, ay ve gün, ardından 24 saatlik dilimde zaman damgası gösterilir.
#### username: Bu alan, Spotify kullanıcı adını gösterir.
#### ms_played: Bu alan, kaç milisaniye çalındığını gösterir.

#### offline_timestamp: Bu alan, çevrimdışı modu kullanıldıysa ne zaman kullanıldığını gösteren zaman damgasıdır.
#### incognito_mode Bu alan, parçanın gizli oturum sırasında çalınıp ("True") çalınmadığını ("False") gösterir.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
import os
```

> Dosya yolumu belirttim. Çünkü normal bir csv dosyası değil. streaming_history klasörümün içinde json dosyaları var.
Her bir json dosyası okunacak ve tabloya dönüşecek. Json dosyalarının içinde gezerek her birini açtım ve dataframe e dönüştürdüm. 
Bütün dosyaları birleştirdim ve csv ye çevirdim.

```python
folder_path = 'C:/Users/Lenovo/Veri Analizi Projelerim/verisetleri/streaming_history'

json_files = [file for file in os.listdir(folder_path) if file.endswith('.json')]

dfs = []
for json_file in json_files:
    file_path = os.path.join(folder_path, json_file)
    
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    df = pd.json_normalize(data)
    dfs.append(df) 

final_df = pd.concat(dfs, ignore_index=True)

final_df.head()

final_df.to_csv('birlesmis_veri.csv', index=False)
```

> Karışık çalma ile şarkı atlama arasındaki ilişkiyi merak ettiğim için bir ısı haritasında görmek istedim. 

```python
pivot_table = final_df.pivot_table(index='Shuffle', columns='Skipped', aggfunc='size', fill_value=0)

sns.heatmap(pivot_table, annot=False, cmap='coolwarm')
plt.title('Shuffle ve Skipped Arasındaki İlişki')
plt.show()
```

> Shuffle False (Kapalı) & Skipped False (Atlanmamış): Bu durum en yüksek sayıya sahip, yani kullanıcıların çoğunluğu shuffle kapalıyken şarkıları atlamamış. > Semicenk'in dinlediğim şarkılarını sıralamak istedim.
Şarkıcı adı Semicenk olan değerleri aldım. Semicenk'i kaç ms dinlediysem onu 3600000'a bölerek saate çevirdim. 

```python
semicenk_tracks = final_df[final_df['Artist Name'] == 'Semicenk']
semicenk_tracks['Hours Played'] = semicenk_tracks['Ms Played'] / 3600000  

top_10_tracks = semicenk_tracks.groupby('Track Name').sum().nlargest(10, 'Hours Played').reset_index()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=top_10_tracks, x='Track Name', y='Hours Played', hue='Track Name', style='Track Name', s=200, palette='viridis')
plt.title('Semi Cenk\'in En Çok Dinlediğim İlk 10 Şarkısı ')
plt.xlabel('Şarkı Adı')
plt.ylabel('Toplam Dinlenme Süresi (Saat)')
plt.xticks(rotation=45)
plt.legend(title='Şarkı Adı', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
yener_tracks = final_df[final_df['Artist Name'] == 'Yener Çevik']
yener_tracks['Hours Played'] = yener_tracks['Ms Played'] / 3600000  

top_3_tracks = yener_tracks.groupby('Track Name').sum().nlargest(10, 'Hours Played').reset_index()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=top_3_tracks, x='Track Name', y='Hours Played', hue='Track Name', style='Track Name', s=200, palette='viridis')
plt.title('Yener Çevik\'in En Çok Dinlediğim İlk 3 Şarkısı')
plt.xlabel('Şarkı Adı')
plt.ylabel('Toplam Dinlenme Süresi (Saat)')
plt.xticks(rotation=45)
plt.legend(title='Şarkı Adı', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df_2024 = final_df[final_df['Year'] == 2024]

hande_yener_tracks = df_2024[df_2024['Artist Name'] == 'Hande Yener']
hande_yener_tracks['Hours Played'] = hande_yener_tracks['Ms Played'] / 3600000 

# Top 20 tracks
top_10_tracks = hande_yener_tracks.groupby('Track Name').sum().nlargest(10, 'Hours Played').reset_index()

# Scatter plot çizimi
plt.figure(figsize=(12, 8))
sns.scatterplot(data=top_10_tracks, x='Track Name', y='Hours Played', hue='Track Name', style='Track Name', s=200, palette='viridis')
plt.title('Hande Yener\'in 2024 Yılında En Çok Dinlediğim İlk 10 Şarkısı ')
plt.xlabel('Şarkı Adı')
plt.ylabel('Toplam Dinlenme Süresi (Saat)')
plt.xticks(rotation=45)
plt.legend(title='Şarkı Adı', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

> Daha güzel bir grafikle aylara da göre toplam dinleme süremi görmek istedim.

```python
final_df['Date'] = pd.to_datetime(final_df['Date'], errors='coerce')

# yıl ve ay sütunu ekledim
final_df['YearMonth'] = final_df['Date'].dt.to_period('M')  # Yıl ve ayı birleştirir, örn: 2022-10

df_grouped = final_df.groupby('YearMonth')['Ms Played'].sum() / 3600000

plt.figure(figsize=(12, 7))
df_grouped.plot(kind='line', marker='o')

plt.title('Aylara Göre Toplam Dinlenme Süresi', fontsize=14)
plt.xlabel('Yıl - Ay', fontsize=12)
plt.ylabel('Toplam Dinlenme Süresi (Saat)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
